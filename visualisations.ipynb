## Downstream Task

```python
import pandas as pd
import matplotlib.pyplot as plt

# New data
data = {
    'Aufgabentyp': ['Text / Code\n generieren', 'Zusammenfassen', 'Fragen \nbeantworten', 'Dialog und \nKonversation', 'Anweisungen \nbefolgen', 'bersetzen'],
    'Vorkommen': [34, 34, 30, 28, 14, 8]
}

df = pd.DataFrame(data)

# Set the color
color = '#3178ca'

# Plot
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Aufgabentyp'], df['Vorkommen'], color=color)

# Adjustments for a nicer plot
plt.xlabel('Aufgabentyp', fontsize=12)
plt.ylabel('Hufigkeit', fontsize=12)
plt.title('Hufigkeit verwendeter Aufgabentypen (Downstream Tasks)', fontsize=14, fontweight='bold')

# Adjust font
plt.xticks(ha='center', fontsize=10)
plt.yticks(fontsize=10)

# Add grid lines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add bar labels
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, str(int(bar.get_height())), ha='center', fontsize=9)

plt.tight_layout()
plt.show()
```
